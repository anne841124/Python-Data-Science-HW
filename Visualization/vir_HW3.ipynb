```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_tips = sns.load_dataset('tips')
data_fmri = sns.load_dataset('fmri')
```

```python
data_tips
```

```python
sns.set(style="darkgrid")

sns.relplot(x = "total_bill", y = "tip", hue = "sex", style = "time"

, data = data_tips);
```

```python
sns.set(style="darkgrid")

sns.relplot(x = "timepoint", y = "signal", hue = "event", style = "region"

, data = data_fmri);
```
