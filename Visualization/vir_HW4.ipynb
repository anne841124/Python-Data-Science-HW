```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data_titanic = sns.load_dataset('titanic')
data_tips = sns.load_dataset('tips')
```

```python
data_titanic.info()
```

```python
data_tips.info()
```

```python
sns.boxplot(data = data_titanic, orient = "h")
plt.show()
```

```python
sns.boxplot(data = data_tips, orient = "h")
plt.show()
```

```python
sns.stripplot(x = "fare", y = "alive", data = data_titanic)
plt.show()
```

```python
sns.stripplot(x = "total_bill", y = "sex", data = data_tips)
plt.show()
```
