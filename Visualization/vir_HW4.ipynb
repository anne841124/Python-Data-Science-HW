```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data_titanic = sns.load_dataset('titanic')
data_tips = sns.load_dataset('tips')
```

```python
data_titanic.info()
```

```python
data_tips.info()
```

```python
sns.barplot(x = 'survived', y = 'age', hue = 'sex', data = data_titanic)
sns.stripplot(x = 'survived', y = 'age', hue = 'sex', data = data_titanic)
plt.show()
```

```python
sns.barplot(x = 'sex', y = 'tip', hue = 'smoker', data = data_tips)
sns.stripplot(x = 'sex', y = 'tip', hue = 'smoker', data = data_tips)
plt.show()
```
