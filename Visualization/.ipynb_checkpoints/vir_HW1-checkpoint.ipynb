```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10 * np.pi, 0.01)
y = np.cos(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
pot_x = np.random.normal(0, 1, 100)
pot_y = np.random.normal(0, 1, 100)
```

```python
plt.scatter(pot_x, pot_y, color = 'red')
plt.show()
```
